```python
C=np.random.randint(0,10,size=(3,3));
D=np.random.randint(0,10,size=(3,3));

Multiplication_of_CD=np.dot(C,D);
Dot_of_CD=C.dot(D);
print("C=")
print(C)
print("D= ")
print(D)
print("Multiplication of CD = ")
print(Multiplication_of_CD);
print("Dot of C.dot(D) = ")
print(Dot_of_CD);
```

Task 3.1.5

```python
B=np.random.randint(0,10,size=(4,4))
A=np.random.randint(0,10,size=(4,4))

result_AdotB= np.dot(A,B)
result_BdotA= np.dot(B,A)

print("A dot B=",result_AdotB)
print("B dot A=",result_BdotA)

if np.array_equal(result_AdotB,result_BdotA):
    print("\nA · B = B · A")
else:
    print("\nA · B ̸= B · A.")
```
